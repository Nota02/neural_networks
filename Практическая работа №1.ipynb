## Задание №1
Напишите функцию, которая делает проверку сразу всех примеров учебного набора с текущими весами, а затем формирует и выводит в консоль получившуюся таблицу истинности.

```python
from random import randint

def generate_weights(n = 2):
  return [randint(1, 100) / 100 for i in range(n)]

w = generate_weights()
w
```

```python
# x - входные данные
# w - веса
def predict(x, w):
  summator = 0
  for i in range(len(w)):
    summator += x[i] * w[i]

  # Пороговая активация
  act = 1 if summator >= 1 else 0

  return act
```

```python
# x - входные данные
# y - истинные данные
# w - веса
def check_dataset(x, y, w, output=True):
  prediction_errors = 0

  if output:
    print("| Входные данные | Предсказание | Ожидаемый вывод")
  for i in range(len(x)):
    predicted = predict(x[i], w)

    prediction_error = y[i] - predicted
    if (prediction_error != 0):
      prediction_errors += 1

    if output:
      print(f"| {x[i]}         | {predicted}            | {y[i]}")

  result = 1 if prediction_errors == 0 else (1 - prediction_errors / len(x))
  return result
```

```python
x_train = [[1,1], [1,0], [0,1], [0,0]]
y_train = [0, 1, 1, 0]

print(f'Точность: {check_dataset(x_train, y_train, w) * 100}%')
```

## Задание №2

Со случайными стартовыми весами обучить нейрон дельта методом для следующих операций:


```python
def train(x_train, y_train, w, SPEED = 0.3):
  for i in range(len(x_train)):
    # Расчет ошибки дельта-методом
    error = y_train[i] - predict(x_train[i], w)

    if error != 0:
      for ii in range(len(w)):
        w[ii] = w[ii] + error * x_train[i][ii] * SPEED
```

```python
def train_and_check(x_train, y_train, MAX_EPOCH_COUNT = 5, speed = 0.3):
  w = generate_weights()
  print(f"Стартовые веса: {w}")

  accuracy = 0
  epoch = 1

  while accuracy < 1 and epoch <= MAX_EPOCH_COUNT:
    train(x_train, y_train, w, speed)
    accuracy = check_dataset(x_train, y_train, w, True)
    print(f"Эпохи: {epoch}, Точность: {accuracy * 100}%, Веса: {w}")

    epoch += 1

  print(f"Конечные веса: {w}")

  return w
```

### Дизъюнкция

```python
train_and_check(
    [[1, 1], [1, 0], [0, 1], [0, 0]],
    [1, 1, 1, 0]
)
```

## Задание №4

Добавить ещё один вход данных на нейрон (на дополнительный вход всегда приходит единица, это аналог нейрона смещения)


```python
def predict(x, w):
    # Добавляем 1 к входным данным (для bias)
    x_with_bias = x + [1]
    summator = 0
    for i in range(len(w)):
        summator += x_with_bias[i] * w[i]
    return 1 if summator > 0 else -1
```

```python
def train_and_check(x_data, y_true, epochs=30, learning_rate=0.1):
    # Инициализация весов
    w = generate_weights(3)

    print(f"Начальные веса: {[round(weight, 3) for weight in w]}\n")

    # Обучение с bias
    for epoch in range(epochs):
        errors = 0
        for x, y in zip(x_data, y_true):
            # Добавляем bias ко входу
            x_with_bias = x + [1]
            prediction = predict(x, w)
            error = y - prediction

            if error != 0:
                errors += 1
                # Обновление всех весов (включая bias)
                for i in range(len(w)):
                    w[i] += learning_rate * error * x_with_bias[i]

        # Вывод прогресса
        if epoch % 10 == 0 or epoch == epochs - 1:
            correct = sum(1 for x, y in zip(x_data, y_true) if predict(x, w) == y)
            accuracy = correct / len(x_data)
            print(f"Эпоха {epoch:3d} | Ошибок: {errors:2d} | Точность: {accuracy*100:.1f}%")

    # Вывод итоговой таблицы
    print("\nИтоговая таблица истинности:")
    print("| Входные данные | Предсказание | Ожидаемый вывод |")
    print("|----------------|--------------|-----------------|")
    for x, y in zip(x_data, y_true):
        prediction = predict(x, w)
        print(f"| {str(x):14} | {prediction:12} | {y:15} |")

    print(f"\nИтоговые веса (w1, w2, bias): {[round(weight, 3) for weight in w]}")
    return w
```

```python
# Данные с двумя входами (bias будет добавлен автоматически)
x_data = [[1, 1], [1, -1], [-1, 1], [-1, -1]]
y_true = [1, -1, -1, -1]
```

```python
final_weights = train_and_check(
    x_data,
    y_true,
    epochs=10,
    learning_rate=0.1
)
```

## Задание №5
Создать и обучить простую нейронную сеть, состоящую из трех нейронов, которая решает задачу XOR (исключающее ИЛИ)

```python
speed = 0.3

# Инициализация весов
w1 = generate_weights()  # нейрон 1 скрытого слоя
w2 = generate_weights()  # нейрон 2 скрытого слоя
w3 = generate_weights()
x_train = [[1, 1], [1, 0], [0, 1], [0, 0]]
y_train = [0, 1, 1, 0]
print(f"Стартовые веса: {w1}, {w2}, {w3}")
```

```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def predict(x, w):
    summator = x[0] * w[0] + x[1] * w[1]
    return sigmoid(summator)
```

```python
def calculate_accuracy(x_train, y_train):
    correct = 0
    for i in range(len(x_train)):
        h1 = predict(x_train[i], w1)
        h2 = predict(x_train[i], w2)
        output = predict([h1, h2], w3)
        if round(output) == y_train[i]:
            correct += 1
    return correct / len(x_train)
```

```python
def train(x_train, y_train, epochs=10000):
    global w1, w2, w3

    for epoch in range(epochs):
        for i in range(len(x_train)):
            x = x_train[i]
            target = y_train[i]

            # Прямое распространение
            h1 = predict(x, w1)
            h2 = predict(x, w2)
            output = predict([h1, h2], w3)

            # Обратное распространение ошибки
            error = target - output

            # Градиенты
            delta_output = error * output * (1 - output)
            delta_h1 = delta_output * w3[0] * h1 * (1 - h1)
            delta_h2 = delta_output * w3[1] * h2 * (1 - h2)

            # Обновление весов
            w3[0] += speed * delta_output * h1
            w3[1] += speed * delta_output * h2

            w1[0] += speed * delta_h1 * x[0]
            w1[1] += speed * delta_h1 * x[1]

            w2[0] += speed * delta_h2 * x[0]
            w2[1] += speed * delta_h2 * x[1]

        if epoch % 1000 == 0:
            total_error = 0
            for i in range(len(x_train)):
                h1 = predict(x_train[i], w1)
                h2 = predict(x_train[i], w2)
                output = predict([h1, h2], w3)
                total_error += (y_train[i] - output) ** 2
            accuracy = calculate_accuracy(x_train, y_train)
            print(f"Epoch {epoch}, Error: {total_error/4:.4f}, Accuracy: {accuracy*100:.2f}%")
```

```python
train(x_train, y_train)
```

```python
print("\nФинальные результаты:")
accuracy = calculate_accuracy(x_train, y_train)
print(f"Финальная точность: {accuracy*100:.2f}%")

print("\nПодробные предсказания:")
for x in x_train:
    h1 = predict(x, w1)
    h2 = predict(x, w2)
    output = predict([h1, h2], w3)
    print(f"Input: {x}, Output: {output:.4f} -> {round(output)} (Expected: {y_train[x_train.index(x)]})")
```
